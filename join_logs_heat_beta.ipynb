```python
def join_one_file(df):
    df['№ пакета'] = df['№ пакета'].apply(lambda x: None if len(str(x))<3 else x)
    df.rename(columns = {'Unnamed: 12': 'время начала т/о', 'Unnamed: 13': 'шаг балок, сек'}, inplace=True)

    df = df.dropna(subset=['время начала т/о', '№ пакета'], how='all').dropna(
        subset=['Номер пакета', '№ пакета'], how='all')
    

    df.reset_index(drop=True, inplace=True)
    
    
    df_left = df.dropna(subset=['марка стали'])
    df_left = df_left[~(df_left['марка стали']=='марка стали')]
    # df_left.to_excel('jule/prepared/tmp2.xlsx')

    df_right = df.copy()

    def right_apply(x):
        try:
            x = int(x)
        except:
            x = None
        return x

    df_right['№ пакета'] = df_right['№ пакета'].apply(right_apply)
    df_right.dropna(subset=['№ пакета'], inplace=True)

    # df_right.to_excel('jule/prepared/tmp_right2.xlsx')

    def add_ind_to_split(df):
        last = df.index[0]-1
        ind = 0
        new_col_ind = []
        for i in df.index:
            if last+1!=i:
                ind+=1
            new_col_ind.append(ind)
            last = i
        df['ind'] = new_col_ind
        return df

    df_right = add_ind_to_split(df_right)
    df_left = add_ind_to_split(df_left)

    part0 = df_right[df_right.ind==0].copy()
    part1 = df_right[df_right.ind==1].copy()

    right0 = part0[pd.isnull(part0['Номер пакета'])]
    centr1 = part0[~pd.isnull(part0['Номер пакета'])]

    right1 = part1[pd.isnull(part1['Номер пакета'])]
    centr2 = part1[~pd.isnull(part1['Номер пакета'])]
    right0.shape, centr1.shape, right1.shape, centr2.shape

    part0 = df_left[df_left.ind==0].copy()
    part1 = df_left[df_left.ind==1].copy()

    centr1_check = part0[~pd.isnull(part0['№ пакета'])]
    left1 = part0[pd.isnull(part0['№ пакета'])]
    
    centr2_check = part1[~pd.isnull(part1['№ пакета'])]
    left2 = part1[pd.isnull(part1['№ пакета'])]

    centr1_check.shape, left1.shape, centr2_check.shape, left2.shape

    left1_for_concat = left1.dropna(axis=1)

    left1_for_concat.columns

    right1_for_concat = right1.dropna(axis=1)

    right1_for_concat.columns

    left1_for_concat.reset_index(inplace=True, drop=True)
    right1_for_concat.reset_index(inplace=True, drop=True)
    try:
        del left1_for_concat['ind']
    except:
        pass
    try:
        del right1_for_concat['ind']
    except:
        pass
    for i in range(50):
        try:
            del right1_for_concat['Unnamed: '+str(i)]
        except:
            pass
        try:
            del left1_for_concat['Unnamed: '+str(i)]
        except:
            pass

    centr_concat = pd.concat([left1_for_concat, right1_for_concat], axis=1)
    # .to_excel('jule/prepared/concat_centr.xlsx')

    ls_need = list(centr_concat.columns)

    ls_need

    right0 = right0[ls_need]
    centr1 = centr1[ls_need]
    print(centr1.columns)
    centr_concat = centr_concat[ls_need]
    print(centr_concat.columns)
    centr2 = centr2[ls_need]
    left2 = left2[ls_need]

    ordered_dfs = [right0, centr1, centr_concat, centr2, left2]
    for i in ordered_dfs:
        print(i.shape)
    return pd.concat(ordered_dfs)
```
