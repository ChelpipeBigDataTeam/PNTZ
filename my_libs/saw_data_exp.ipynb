```python
def lowerstr(f):
    """Приводим все символы в номерах партий к нижнему регистру"""
    for col in f:
        if '№ партии' in col:
            f['№ партии'] = f['№ партии'].apply(lambda x: str(x))            
            f['№ партии'] = f['№ партии'].str.lower()
        if '№ плавки' in col:
            f['№ плавки'] = f['№ плавки'].str.lower()
    return f
```

```python
import pandas as pd
import numpy as np
import warnings
import re
from collections import OrderedDict
warnings.filterwarnings('default')

pd.options.display.max_columns = 100
pd.options.display.max_rows = 1000
```

```python
# подгрузка всех файлов
import os
dr = '../DATA/source/exp'
file_names = os.listdir(dr)
```

```python
file_names.remove('prepared')
```

```python
# файлы в словарь
df_dicts = {}
for name in file_names:
    df_dicts[name] = pd.read_excel(dr+'\\'+name, None)
    print(name)
```

```python
# список всех таблиц из всех файлов
sheetname_list = []
for name, dict_ in zip(df_dicts, df_dicts.values()):    
    tmp = dict_.keys()
    tmp = list(tmp)
    tmp = tmp[0:-3]
    sheetname_list.append(tmp)
len(sheetname_list)
```

```python
# dict_sheet_count - словарь всех вкладок(листов) из таблиц без дубликатов и их частоты
# uniq_sheetname - множество всех вкладок(листов) из таблиц без дубликатов (не используется)

dict_sheet_count = {}
uniq_sheetname = set()
for st in sheetname_list:
    for ls in st:
        uniq_sheetname.add(ls)


uniq_sheetname = list(uniq_sheetname)
uniq_sheetname.sort()
print(len(uniq_sheetname))

# добавление частот в словарь
for st in sheetname_list:
    for ls in st:
        counts = dict_sheet_count.get(ls,0)
        dict_sheet_count[ls] = counts + 1
        
        
# df - для вывода словаря в excel            
df = pd.DataFrame.from_dict(data = dict_sheet_count, orient='index')
```

```python
def find_sheets_for_part(part_number):
    sheet_names = []
    file_names = []
    for name, df_dict in zip(df_dicts, df_dicts.values()):
        for sheet_name, sheet in zip(df_dict, df_dict.values()):
            if '№ партии' in sheet:
                if sheet[sheet['№ партии']==part_number].shape[0]>0:
                    print(sheet_name, sheet[sheet['№ партии']==part_number].shape[0], '_________',name)
                    file_names.append(name)
                    sheet_names.append(sheet_name)
    sheet_names= list(set(sheet_names))
    file_names = list(set(file_names))
    return file_names, sheet_names
```

```python
def find_rows(file_names, sheet_names, part_number):
    df = pd.DataFrame()
    for file_name in file_names:
        for sheet_name in sheet_names:
            if sheet_name in df_dicts[file_name]:
                df_tmp = df_dicts[file_name][sheet_name]
                df_tmp = df_tmp[df_tmp['№ партии']==part_number]
                df = pd.concat([df, df_tmp])
    df_dicts[file_name][sheet_name]
    return df
```
